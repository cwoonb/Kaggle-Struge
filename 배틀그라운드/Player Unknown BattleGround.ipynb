```python
def clean_name(name):
    return name.strip().lower().replace(" ", "_")
df.rename(columns=clean_name, inplace = True)
```

```python
# Different Weapon Type
sns.set()
plt.figure(figsize=(20,5))
sns.countplot (df["weapon_type"], palette="rainbow")
plt.show()
# Most occuring assualt Rifle
```

## 가장 중요한 것은 매거진 용량의 크기입니다
PUBG 또는 CSGO와 같은 TPS (Third Person Shooter) 게임을 플레이하는 동안 항상 좋은 매거진 크기의 총이 필요합니다. 연속 탄환으로 장거리에서 플레이어를 쏘는 데 도움이 될 것입니다.
적과의 근접 전투 또는 1 대 1 전투 중에 대형 탄창이 총을 뿌려 적을 죽이는 데 도움이됩니다.

```python
sns.set()
plt.figure(figsize = (10,10))
sns.countplot (df.magazine_capacity, palette="inferno")
plt.show()

# Most commonly a gun has 30 bullets 
```

## 또 다른 중요한 측면은 "총의 손상"입니다.

특정 무기의 총알이 발사되면 플레이어에 대한 피해량을 결정합니다

```python
temp = df.groupby("weapon_type")["damage"].count().reset_index()
```

```python
# The weapon damage according to the size of rifle
plt.figure(figsize = ( 6,6))
sns.barplot(y="weapon_type", x ="damage",data = temp)
plt.show()
```

## 여기에서 돌격 소총의 데미지가 더 높다는 결론을 내릴 수 있습니다.
1. AKM, M416, SCAR-L 등
그러나 가장 높은 데미지를 가진 소총 ??? 더 분석해보자  

```python
most_damaging_weapon = df[df["weapon_type"] == "Assault Rifle"][["damage", "weapon_name"]].sort_values("damage", ascending  = False)
most_damaging_weapon
```

```python
plt.figure(figsize = (6,6))
sns.barplot(y="weapon_name", x ="damage",data = most_damaging_weapon)
plt.show()
```

## 평균 총알 속도

```python
speed = df.groupby("weapon_type")["bullet_speed"].mean().reset_index().dropna()
speed
```

```python
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)
ax.vlines(x=speed["weapon_type"], ymin=0, ymax=speed["bullet_speed"], color='firebrick', alpha=0.7, linewidth=2)
ax.scatter(x=speed["weapon_type"], y=speed["bullet_speed"], s=75, color='firebrick', alpha=0.7)
#Annotate Title
ax.set_title('무기 유형별 평균 총알 속도', fontdict={'size':22})
ax.set_ylabel('Average Speed ')
ax.set_xticks(speed.weapon_type)
plt.show()
```

## 경량 기관총은 단위 면적당 총알의 속도가 빠릅니다.

## 이제 각 무기 유형의 평균 발사 속도를 분석하겠습니다.

```python
fire_rate = df.groupby("weapon_type")["rate_of_fire"].mean().reset_index().dropna()

fire_rate
```

```python
# Average Fire Rate
plt.figure(figsize=(20,5))
plt.title("
무기 유형별 평균 발사 속도")
sns.barplot (x = "weapon_type", y = "rate_of_fire", data = fire_rate, palette="rainbow")
plt.show()
```


돌격 소총의 발사 속도가 가장 낮음을 분명히 알 수 있습니다.

```python
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['figure.dpi'] = 150
sns.scatterplot(x=df.bullet_speed, y = df.damage)
plt.show()
```


기본 삽입과 분석을 통해 총알 속도의 증가가 더 높은 피해를 입히는 것으로 나타났습니다
